```python
import pandas as pd
import numpy as np
import pickle
from catboost import CatBoostClassifier
from sklearn.preprocessing import MinMaxScaler

# 1. Setup Feature Names
feature_names = [
    "Tenure", "WarehouseToHome", "NumberOfDeviceRegistered", "PreferedOrderCat",
    "SatisfactionScore", "MaritalStatus", "NumberOfAddress", "Complain",
    "DaySinceLastOrder", "CashbackAmount"
]
scale_vars = ["Tenure", "WarehouseToHome", "NumberOfDeviceRegistered", "SatisfactionScore", "NumberOfAddress"]

# 2. Create Dummy Data to initialize the model and scaler
np.random.seed(42)
X_dummy = pd.DataFrame(np.random.rand(100, 10), columns=feature_names)
y_dummy = np.random.randint(0, 2, 100)

# 3. Create and Save Scaler
scaler = MinMaxScaler()
scaler.fit(X_dummy[scale_vars])
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

# 4. Create and Save CatBoost Model
model = CatBoostClassifier(iterations=10, depth=3, learning_rate=0.1, verbose=False)
model.fit(X_dummy, y_dummy)
with open('best_model.pkl', 'wb') as f:
    pickle.dump(model, f)

# 5. Create Feature Importance CSV for the bar chart
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Feature Importance Score': model.get_feature_importance()
})
importance_df.to_csv("data_ecommerce_customer_churn_importance.csv", index=False)

print("✅ Success! Files created: 'best_model.pkl', 'scaler.pkl', and 'data_ecommerce_customer_churn_importance.csv'")
```
